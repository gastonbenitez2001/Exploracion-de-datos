```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
```

```python
def distribucion_normal(x,mu,sigma):

    return norm.pdf(x,mu,sigma)
```

```python
# El eje x oscila entre -3 y 3 con 0,001 pasos
x = np.arange(-3, 3,0.0001)

#Valores de mu y sigma de la distribucion normal
mu = 0 #--> Media
sigma = 1 #--> Varianza

#Valores de mu y de sigma usados para ajustar parametros
mu_estimacion = 1
sigma_estimacion = 1

#Implementacion de PLOT
plt.plot(x,distribucion_normal(x,mu,sigma)) #--> Distribucion normal tipica
plt.show()

plt.plot(x,distribucion_normal(x,mu_estimacion,sigma_estimacion))
plt.show()
```
